```python
IPA_CHARS = ['p', 'e', 'ɾ', 'o', 'i', 's', 'x', 'a', 'l', 'k', 'd', 'n', 'we',
             'b', 't', 'm', 'ɡ', 'f', 'u', 'jo', 't͡ʃ', 'ja', 'ʝ', 'je', '',
             'si', 'se', 'xe', 'wa', 'w', 'ɡi', 'r', 'ks', 'ɡw', 'wi', 'ɡe',
             'xi', 'ju', 'wo', 'ʃ', 'ɲ', 't͡ɬ', 't͡s', 'æ', 'ɹ̩', 'ð', 'ɪ', 'j',
             'd͡ʒ', 'ʌ', 'ɑ', 'ə', 'h', 'ɹ', 'ŋ', 'ɔ', 'ʊ', 'z', 'v', 'ɛ', 'θ',
             'ʒ']
ABC = 'abcdefghijklmnopqrstuvwxyz'
SPANISH_ABC = 'abcdefghijklmnopqrstuvwxyzñúóéáíü'

english = stanza.Pipeline(lang='en', processors='tokenize,lemma')
spanish = stanza.Pipeline(lang='es', processors='tokenize,lemma')
trans_spanish = epitran.Epitran('spa-Latn')
trans_english = epitran.Epitran('eng-Latn')
langs = {Language.ENGLISH: 'en', Language.SPANISH: 'es', None: 'none'}
detector = LanguageDetectorBuilder.from_languages(Language.ENGLISH, Language.SPANISH).build()


def is_lemma(word: str, lang: str) -> str:
    lemma = ''
    if lang == ENG and langs[detector.detect_language_of(word)] == 'en':
        lemma = english(word).sentences[0].words[0].lemma
        if not lemma:
            return ''
        lemma = lemma.lower()
        ipa = trans_english.trans_list(lemma)
        for char in ipa:
            if char not in IPA_CHARS:
                print('no ipa', word, lemma)
                return ''
        for char in lemma:
            if char not in ABC:
                print('no alphabet', word, lemma)
                return ''
    elif lang == SPA and langs[detector.detect_language_of(word)] == 'es':
        lemma = spanish(word).sentences[0].words[0].lemma
        if not lemma:
            return ''
        lemma = lemma.lower()
        ipa = trans_spanish.trans_list(lemma)
        for char in ipa:
            if char not in IPA_CHARS:
                print('no ipa', word, lemma)
                return ''
        for char in lemma:
            if char not in SPANISH_ABC:
                print('no alphabet', word, lemma)
                return ''
    if word == lemma:
        return word
    else:      
        return ''


def keep_lemmas(word_vec_file: str, lang: str, output_file: str, from_=0) -> None:
    with open(word_vec_file, 'r') as f:
        with open(output_file, 'w') as ff:
            for i, line in enumerate(f):
                if i < from_:
                    continue
                split_line = line.split()
                word, vec = split_line[0], split_line[1:]
                if len(vec) == 300 and is_lemma(word, lang):
                    ff.write(line)
                if i % 5000 == 0:
                    print(i, word, is_lemma(word, lang))
```

```python
keep_lemmas(ROOT + 'embeddings/wiki.en.align.vec', 'eng', ROOT + 'embeddings/en_embeddings_new.vec')
```

```python
def keep_only_lang(vec_path: str, lex_path: str, new_vec_path: str) -> None:
    lexicon = list()
    count = 0
    with open(lex_path, 'r') as f:
        for line in f.readlines():
            line = line.split()
            lexicon.append(line[0].lower())
    lexicon = list(set(lexicon))
    c = 0
    with open(vec_path, 'r') as f:
        with open(new_vec_path, 'w') as ff:
            for line in f.readlines():
                c += 1
                split_line = line.split()
                if split_line[0] in lexicon:
                    count += 1
                    ff.write(line)
                if c % 10000 == 0:
                    print(c)
    print(f'Number of words in lexicon: {len(lexicon)}')
    print(f'Total word vectors: {count}')
```

```python
keep_only_lang(ROOT + 'embeddings/en_embeddings_1.vec', 
               ROOT + 'embeddings/lexicons/english_words.txt',
               ROOT + 'embeddings/en_embeddings_2.vec')
```
